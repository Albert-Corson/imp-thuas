```python
import pandas as pd
import numpy as np

targets = [
    "Relative atmospheric humidity",
    "Temperature",
    "Global Radiation",
    "smartMeter_power",
    "alklimaHeatPump_flow_temp",
    "alklimaHeatPump_op_mode",
    "co2sensor_co2",
]

col = [
    "Root mean squared error",
    "Average variance error",
]

for target in targets:
    print(target)
    try:
        df = pd.read_csv(f"saved_results/Results - {target}.csv", sep=';', index_col=0)
        df = df[df.index.str.contains("Hot|RNN|forward|KNN", regex=True)]
        df["Root mean squared error"] = np.sqrt(df["Mean squared error"].values)
        for i in range(1, 6):
            subdf = df[col][df.index.str.contains(f"gap type {i}")].copy()
            subdf.sort_values(col, inplace=True)
            if len(subdf.index) == 0:
                continue
            pcols = []
            for column in col:
                pcols.append(column + " (%)")
                subdf[pcols[-1]] = np.round((subdf[column].values / subdf[column].values.min() * 100) - 100, 2)
            subdf["Magical score"] = subdf[pcols].sum(axis=1)
            subdf.sort_values(["Magical score"], inplace=True)
            display(subdf)
    except:
        pass
```
